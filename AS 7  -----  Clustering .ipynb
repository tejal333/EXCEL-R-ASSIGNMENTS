# Hierarchy clustring

```python
plt.figure(figsize = (16,10));
dendrogram=sch.dendrogram(sch.linkage(a,method='average'))
```

```python
# Agglomerative Clustering
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'average')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(a)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
airlines1 = pd.concat([a,Clusters], axis=1)
airlines1
```

```python
airlines1.sort_values("Clusters")
```

K-MEANS

```python
kmeans = KMeans(n_clusters=14,random_state=0)
kmeans.fit(a)
```

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(1, 19):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(a)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1, 19), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
